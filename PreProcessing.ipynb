```python
# Removing incomplete tuples

data.dropna(axis=0, how='any', inplace=True)
```

```python
# Converting Positions

forward = ['RF', 'LF', 'ST', 'RS', 'LS', 'CF', 'LW', 'RW']
mid = ['RCM', 'LCM', 'LDM', 'CAM', 'CDM', 'RM', 'LAM', 'LM', 'RDM', 'CM', 'RAM']
defender = ['RCB', 'CB', 'LCB', 'LB', 'RB', 'RWB', 'LWB']

data['Position'].replace(['RF', 'LF', 'ST', 'RS', 'LS', 'CF', 'LW', 'RW'], 'Forward', inplace=True)
data['Position'].replace(['RCM', 'LCM', 'LDM', 'CAM', 'CDM', 'RM', 'LAM', 'LM', 'RDM', 'CM', 'RAM'], 'MidFielder', inplace=True)
data['Position'].replace(['RCB', 'CB', 'LCB', 'LB', 'RB', 'RWB', 'LWB'], 'Defender', inplace=True)
```

```python
data.isnull().sum()
```

```python
heightConv = [30.48, 2.54]
data['Height'] = data['Height'].apply(lambda x: round(pd.Series(map(int, x.split('\''))).dot(heightConv), 1))

weightConv = [0.454]
data['Weight'] = data['Weight'].str.strip('lbs').astype(float)
data['Weight'] = data['Weight'].apply(lambda x: round(pd.Series(x).dot(weightConv), 1))
```

```python
data['Value'] = data['Value'].str.strip('€')

for i in range(data.shape[0]):
    temp = str(data['Value'].iloc[i])
    if temp[-1] == 'K':
        temp = temp[:-1]
    elif temp[-1] == 'M':
        temp = temp[:-1]
        temp = float(temp) * 1000
    data['Value'].iloc[i] = temp
data['Value'] = data['Value'].astype(float)
```
